# 考試題目

## 因為武漢肺炎, 所在國家執行全境封鎖180天
## 因應全境封鎖需要購買大量食物
## 你的家庭共有三人, 每日需要熱量6500卡路里
## 請達成以下三個目標
## 1. 計算每樣食物需要採購的數量
## Target_A. 食物數量＊卡路里 ~ 接近1,170,000
## Target_B. 食物數量＊幸福感 ~ 接近8
## Target_C. 所需金額為相對 ~ 接近150,000 NTD
## 2. 最終函數比重為= A*0.35 +B*0.35 + C*0.3

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
from sklearn import metrics
import random
import string
from decimal import getcontext, Decimal
```

```python
Food_index = pd.read_csv("Food_index.csv", encoding='big5hkscs')
Food_index.head()
```

```python
Target = [1170000, 8, 150000]
Calories = (225, 660, 1400, 900, 403, 80, 63, 136, 225, 630)
Happy = (5, 9, 8, 7, 2, 6, 10, 10, 7, 6)
Money = (10, 157, 160, 134, 30, 5, 25, 25, 38, 38)
Weight = (0.35, 0.35, 0.3)
Columns = ["飯/麵(碗)", "牛排", "牛肉", "豬肉", "泡麵(碗)", "蛋(顆)", "咖啡(杯)", "可樂(杯)", "牛奶", "餅乾"]
Control_table = pd.DataFrame()
```

```python
# 產生一筆購買清單
def generate_seed():
    x = (
        random.randint(180, 540),
        random.randint(20, 859),
        random.randint(15, 835), 
        random.randint(15, 1007),
        random.randint(60, 2903),
        random.randint(100, 14625),
        random.randint(30, 90),
        random.randint(30, 90),
        random.randint(20, 90),
        random.randint(20, 60)
    )
    return x
```

```python
# 產生 100 筆購買清單
def seed_table():
    buy_count = []
    for i in range(100):
        buy_count.append(generate_seed())
    return buy_count
```

```python
# 取得最終分數
def get_finalpreds(df):
    df_a, df_b, df_c = df, df, df
    
    df_a = df_a * Calories
    df_b = df_b * Happy
    df_c = df_c * Money
    
    df_t = pd.DataFrame([df_a.sum(axis=1), df_b.sum(axis=1), df_c.sum(axis=1)])
    df_t = df_t.T
    df_t.columns = ['Target_a', 'Target_b', 'Target_c']
    
    df_t = np.abs(df_t - Target) / Target
    df_t = df_t * Weight
    
    f = np.abs(df_t.mean(axis=1))
    f = pd.DataFrame(f)
    f.columns = ['Final_preds']
    
    return f
```

```python
df1 = pd.DataFrame(seed_table())
df1.columns = Columns
```

```python
df1.head()
```

```python
for count in range(5):
    
    if count == 0:
        # 算出最終分數
        finalscore = get_finalpreds(df1)

        # 合併到DataFrame並做出排名
        df1 = pd.concat([df1, finalscore], axis=1)
        df1 = df1.sort_values(by=['Final_preds'], ascending=[True])
        df1 = df1.reset_index(drop=True)

        # 儲存到Control_table
        Control_table = pd.concat([Control_table, df1])
        Control_table = Control_table.sort_index(ascending=[True])
        pass
    else :
        # 從Control_table取出前100名
        df1 = Control_table[:100]
        df1 = df1.drop(columns=['Final_preds'], axis=1)
        # 複製一倍
        df1 = pd.concat([df1, df1])

        # 隨機座標 菁英保留前三比
        x = []
        for i in range(10):
            x.append([random.randint(3, 199), random.randint(0, 9)])

        # 突變
        for i in x:
            df_mu = pd.DataFrame(generate_seed())
            df_mu = df_mu.T

            df1.iloc[i[0], i[1]] = df_mu.iloc[0, i[1]]

        # 交配
        for i in range(10):
            x_rand = random.randint(3, 199)
            x_rand2 = random.randint(3, 199)
            y_rand = random.randint(0, 9)

            c = df1.iloc[x_rand, y_rand]
            df1.iloc[x_rand, y_rand] = df1.iloc[x_rand2, y_rand]
            df1.iloc[x_rand2, y_rand] = c

        df1 = df1.reset_index(drop=True)

        # 算出最終分數
        finalscore = get_finalpreds(df1)

        # 合併到DataFrame並做出排名
        df1 = pd.concat([df1, finalscore], axis=1)
        df1 = df1.sort_values(by=['Final_preds'], ascending=[True])
        df1 = df1.reset_index(drop=True)

        # 儲存到Control_table
        Control_table = pd.concat([Control_table, df1[:100]])
        Control_table = Control_table.sort_values(by=['Final_preds'], ascending=[True])
        Control_table = Control_table.reset_index(drop=True)
```

```python
Control_table.to_csv('Final.csv', index=False)
```
