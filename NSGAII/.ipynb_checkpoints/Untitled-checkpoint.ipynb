```python
#generate 100 seed
all_quantity = []
for i in range(100):
    quantity = (
    random.randint(180,540),
    random.randint(20,26),
    random.randint(15,20),
    random.randint(15,20),
    random.randint(60,83),
    random.randint(100,137),
    random.randint(30,90),
    random.randint(30,90),
    random.randint(20,90),
    random.randint(20,60))
    all_quantity.append(quantity)
print(all_quantity)
```

```python
#check length
len(all_quantity)
```

```python
#add to frame
quantity_frame = pd.DataFrame(all_quantity)
quantity_frame.columns = ['飯/麵(碗)','牛排','牛肉','豬肉','泡麵(碗)','蛋(顆)','咖啡(杯)','可樂(杯)','牛奶','餅乾']
quantity_frame.head()
```

```python
#Set your target
target_frame =[]
target_frame = pd.DataFrame(target_frame)

target_frame['SumOfCalorie'] = (quantity_frame['飯/麵(碗)'] + quantity_frame['牛排'] + quantity_frame['牛肉']+ quantity_frame['豬肉']+ quantity_frame['泡麵(碗)']+ quantity_frame['蛋(顆)']+ quantity_frame['咖啡(杯)']+ quantity_frame['可樂(杯)']+ quantity_frame['牛奶']+ quantity_frame['餅乾'])*4722
target_frame['SumOfHappiness'] = (quantity_frame['飯/麵(碗)'] + quantity_frame['牛排'] + quantity_frame['牛肉']+ quantity_frame['豬肉']+ quantity_frame['泡麵(碗)']+ quantity_frame['蛋(顆)']+ quantity_frame['咖啡(杯)']+ quantity_frame['可樂(杯)']+ quantity_frame['牛奶']+ quantity_frame['餅乾'])*(5+9+8+7+2+6+10+10+7+6)
target_frame['SumOfMoney'] = (quantity_frame['飯/麵(碗)'] + quantity_frame['牛排'] + quantity_frame['牛肉']+ quantity_frame['豬肉']+ quantity_frame['泡麵(碗)']+ quantity_frame['蛋(顆)']+ quantity_frame['咖啡(杯)']+ quantity_frame['可樂(杯)']+ quantity_frame['牛奶']+ quantity_frame['餅乾'])*(10+157+160+134+30+5+25+25+38+38)
```

```python
target_frame.head()
```

```python
calorie_target = 1170000
happiness_target = 8
money_target = 150000
```

```python
calorie_final_target=np.abs(target_frame['SumOfCalorie'] - calorie_target)
happiness_final_target=np.abs(target_frame['SumOfHappiness'] - happiness_target)
money_final_target=np.abs(target_frame['SumOfMoney'] - money_target)
```

```python
calorie_final_target.head()
#happiness_final_target.head()
#money_final_target.head()
```

```python
calorie_percentage_frame=calorie_final_target/calorie_target
happiness_percentage_frame=happiness_final_target/happiness_target
money_percentage_frame=money_final_target/money_target
```

```python
#calorie_percentage_frame.head()
happiness_percentage_frame.head()
#money_percentage_frame.head()
```
