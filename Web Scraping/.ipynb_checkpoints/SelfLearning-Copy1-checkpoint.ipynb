```python
from bs4 import BeautifulSoup as soup
import requests


my_url = requests.get("https://www.books.com.tw/web/sys_bbotm/books/010101/?o=1&v=1&page=")
my_url.encoding='utf8'
books = []
```

```python
page_soup = soup(my_url.text, "html.parser")
```

```python
containers = page_soup.find("div",{"class":"wrap"})
```

```python
print(containers)
```

```python
container = containers.find_all("div",{"class":"item"})
for items in container:
    title = items.find("h4").text
    info = items.find("li").text.split("，")
    authors = info[0]
    publisher = info[1]
    date = info[2].split("：")[1]
    price = items.find("div",{"class":"price_box"}).li.text.strip("優惠價：")
    link = items.find("a")["href"]
    
    print(title)
    print(authors)
    print(publisher)
    print(date)
    print(price)
    print(link)
    
    books.append({"Author":authors,"Title":title,"Publisher":publisher,"Date":date,"Price":price,"Link":link})
```

```python
import pandas as pd

# Create a Pandas dataframe from the data.
df1 = pd.DataFrame(books)

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer1 = pd.ExcelWriter('Books.xlsx', engine='xlsxwriter')#改檔名
# Convert the dataframe to an XlsxWriter Excel object.
df1.to_excel(writer1, sheet_name='Sheet1')
# Close the Pandas Excel writer and output the Excel file.
writer1.save()
```
