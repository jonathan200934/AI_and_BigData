```python
from bs4 import BeautifulSoup as soup
import requests

my_url = requests.get("https://www.newegg.com/global/tw-en/p/pl?N=101499820&ActiveSearchResult=True")
my_url.encoding='utf8'
```

```python
page_soup = soup(my_url.text, "html.parser")
```

```python
containers = page_soup.find_all("div",{"class":"item-container"})
```

```python
filename = "product.csv"
f = open(filename, "w")

headers = "brands, product_name, prices\n"

f.write(headers)
```

```python
for item in containers:
    brands = item.find("div",{"class":"item-branding"}).a.img["title"]
    
    title_container = item.find_all("a",{"class":"item-title"})
    product_name = title_container[0].text
    
    prices = item.find("li",{"class":"price-current"}).text.strip()
    
    print("brands: " + brands)
    print("product_name: " + product_name)
    print("prices: " + prices)
    
    f.write(brands.strip("?V") + "," + product_name.replace(",", "|") + "," + prices.replace("\xa0","|"))
    
f.close()
```
